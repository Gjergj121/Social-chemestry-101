```python
y = [0.313, 0.396, 0.318, 0.418, 0.316, 0.428]

#plt.yticks(y) # to show the exact y-values  
x = ["Flant5-small - no ROT categories", "Flant5-small - with ROT categories", "Flant5-base - no ROT categories", "Flant5-base - with ROT categories", "Flant5-large - no ROT categories", "Flant5-large - with ROT categories"]
#plt.xticks(x)

y = y[::-1]
x = x[::-1]

plt.figure(figsize=(15, 8), facecolor='lightsteelblue')

plt.barh(x[0], y[0], color="royalblue")
plt.barh(x[1], y[1], color="navajowhite")
plt.barh(x[2], y[2], color="royalblue")
plt.barh(x[3], y[3], color="navajowhite")
plt.barh(x[4], y[4], color="royalblue")
plt.barh(x[5], y[5], color="navajowhite")


for i, v in enumerate(y):
    if i % 2 == 0:
        plt.text(v, i, str(v), color='forestgreen', fontweight='bold', ha='left', va='center', fontsize=13)
    else:
        plt.text(v, i, str(v), color='firebrick', fontweight='bold', ha='left', va='center', fontsize=13)

#plt.ylabel('Models', weight="bold")
#plt.xlabel('Occurrences', labelpad=1)
plt.title('Evaluation using Bleu-1', weight="bold", fontsize=20)

plt.tick_params(axis='y', labelsize='large', pad=5)

plt.yticks(fontweight="bold", fontsize=20)
plt.xticks(fontweight="bold", fontsize=15)

plt.show()
```
